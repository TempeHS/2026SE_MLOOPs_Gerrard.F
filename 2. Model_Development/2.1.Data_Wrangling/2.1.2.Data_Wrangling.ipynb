### Remove Outliers

Removing `Outliers` is essential for my dataset as for a couple of features there is a single outlier that is drastically out of range. When I testing the compatibility between my features earlier on, I found the correlation was surprisingly low despite the logical relationship that was clearly visible. I found that a specific outlier may potentially be missing the `local trend` as it forces the `matplotlib` to account for the noise.

Thus I must remove these outliers, to improve my `model's accuracy`.

```python
# I'm rounding so the numbers are more interpretable for both the machine and I
# Identify IQR for kills_total to highlight outliers
Q1 = data_frame["kills_total"].quantile(0.25)
Q3 = data_frame["kills_total"].quantile(0.75)
IQR = Q3 - Q1
print(
    f"Outliers are a kills_total above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}"
)
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["kills_total"] >= Q1 - 1.5 * IQR)
    & (data_frame["kills_total"] <= Q3 + 1.5 * IQR)
]

# Identify IQR for wins_total to highlight outliers
Q1 = data_frame["wins_total"].quantile(0.25)
Q3 = data_frame["wins_total"].quantile(0.75)
IQR = Q3 - Q1
print(
    f"Outliers are a wins_total above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}"
)
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["wins_total"] >= Q1 - 1.5 * IQR)
    & (data_frame["wins_total"] <= Q3 + 1.5 * IQR)
]

# Identify IQR for matchs_played to highlight outliers
Q1 = data_frame["matchs_played"].quantile(0.25)
Q3 = data_frame["matchs_played"].quantile(0.75)
IQR = Q3 - Q1
print(
    f"Outliers are a matchs_played above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}"
)
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["matchs_played"] >= Q1 - 1.5 * IQR)
    & (data_frame["matchs_played"] <= Q3 + 1.5 * IQR)
]

# Identify IQR for k/d to highlight outliers
Q1 = data_frame["k/d"].quantile(0.25)
Q3 = data_frame["k/d"].quantile(0.75)
IQR = Q3 - Q1
print(
    f"Outliers are a k/d above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}"
)
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["k/d"] >= Q1 - 1.5 * IQR) & (data_frame["k/d"] <= Q3 + 1.5 * IQR)
]

# Identify IQR for dmg/rnd to highlight outliers
Q1 = data_frame["dmg/rnd"].quantile(0.25)
Q3 = data_frame["dmg/rnd"].quantile(0.75)
IQR = Q3 - Q1
print(
    f"Outliers are a dmg/rnd above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}"
)
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["dmg/rnd"] >= Q1 - 1.5 * IQR)
    & (data_frame["dmg/rnd"] <= Q3 + 1.5 * IQR)
]

# Identify IQR for win% to highlight outliers
Q1 = data_frame["win%"].quantile(0.25)
Q3 = data_frame["win%"].quantile(0.75)
IQR = Q3 - Q1
print(
    f"Outliers are a win% above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}"
)
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["win%"] >= Q1 - 1.5 * IQR) & (data_frame["win%"] <= Q3 + 1.5 * IQR)
]

# Identify IQR for headshot_% to highlight outliers
Q1 = data_frame["headshot_%"].quantile(0.25)
Q3 = data_frame["headshot_%"].quantile(0.75)
IQR = Q3 - Q1
print(f"Outliers are a headshot_% above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}")
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["headshot_%"] >= Q1 - 1.5 * IQR)
    & (data_frame["headshot_%"] <= Q3 + 1.5 * IQR)
]

# Identify IQR for avg_kill_per_rnd to highlight outliers
Q1 = data_frame["avg_kill_per_rnd"].quantile(0.25)
Q3 = data_frame["avg_kill_per_rnd"].quantile(0.75)
IQR = Q3 - Q1
print(
    f"Outliers are a avg_kill_per_rnd above {(Q3 + IQR * 1.5):.2f} or below {(Q1 - IQR * 1.5):.2f}"
)
# Adjust range to IQR
data_frame = data_frame[
    (data_frame["avg_kill_per_rnd"] >= Q1 - 1.5 * IQR)
    & (data_frame["avg_kill_per_rnd"] <= Q3 + 1.5 * IQR)
]
```

### Scale Features to adjusted range

By scaling the range of features it helps the model better understand the trend as the scale of each individual feature won't impact the model's understanding of features with smaller ranges. In my case, kills_total shouldn't completely outshine win% as the difference between their ranges is apparent but specific to each features definition.

The model I plan to use is a multiple feature linear regression, so scaling is extremely important. Some features that I hope to use may completely lose their value as due to the drastic difference in ranges, the model will favour larger ranges increasing their weighting on the trend. This already introduces the first statistical bias in the model.

For logical purposes some statistics can not be negative thus when I scale the minimum will have to be 0 rather than the provided range.

```python
# Minimum value with space for outliers
MIN_headshot = 0
MIN_matchs_played = 0
MIN_kills_total = 0
MIN_wins_total = 0
MIN_kd = 0 
MIN_dmgrnd = 80
MIN_avg_kill_per_rnd = 0 


# Maximum value with space for outliers
MAX_headshot = 32
MAX_matchs_played = 1350
MAX_kills_total = 24000
MAX_wins_total = 700
MAX_kd = 2
MAX_dmgrnd = 200
MAX_avg_kill_per_rnd = 1.25

# Scale Features
data_frame["k/d"] = (data_frame["k/d"] - MIN_kd) / (MAX_kd - MIN_kd)

data_frame["matchs_played"] = (data_frame["matchs_played"] - MIN_matchs_played) / (MAX_matchs_played - MIN_matchs_played)

data_frame["kills_total"] = (data_frame["kills_total"] - MIN_kills_total) / (
    MAX_kills_total - MIN_kills_total
)

data_frame["wins_total"] = (data_frame["wins_total"] - MIN_wins_total) / (
    MAX_wins_total - MIN_wins_total
)

data_frame["headshot_%"] = (data_frame["headshot_%"] - MIN_headshot) / (MAX_headshot - MIN_headshot)

data_frame["dmg/rnd"] = (data_frame["dmg/rnd"] - MIN_dmgrnd) / (MAX_dmgrnd - MIN_dmgrnd)

data_frame["avg_kill_per_rnd"] = (
    data_frame["avg_kill_per_rnd"] - MIN_avg_kill_per_rnd
) / (MAX_avg_kill_per_rnd - MIN_avg_kill_per_rnd)

data_frame.describe()
```
