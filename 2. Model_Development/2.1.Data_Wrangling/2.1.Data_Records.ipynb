### New Features

| Feature | Calculation | Purpose |
| - | - | - |
| performance | `(data_frame["dmg_indicator"] * data_frame["consistency"]) * 10` | To measure efficiency of damage in accordance to your efficiency thus will cancel out the bias of experience and won't favour players who have played longer |
| skill | `(data_frame['headshot_%'] * data_frame['avg_kill_per_rnd'])` | Hitting headshots are one of the hard habits in valorant thus not many people prioritise it. As it may not be a clear indicator of skill, a single bullet to kill is definitely much more efficient in comparison to spraying. This definitely does have bias in the sense that players who may have killed one person with a headshot will differ in comparison to a person with 10 kills with no headshots, despite the clear kill difference. |
| consistency | `data_frame["wins_total"] / data_frame["matchs_played"]` | A player's ability to be perform at the same level constantly, thus establishing your ability to perform at a high or low rate |
| dmg_indicator | `(data_frame['k/d'] * data_frame["dmg/rnd"]).round(2)` | A simple amplification of two variables that relate to one another, this to create a foundation for a number that I can manipulate|
