## Split Data into Training and Testing set

Split aggregate dataset into a training and testing set to allow the machine to test its efficiency on unseen data. 

### Import dependencies

```python
# Import frameworks
import pandas as pd
from sklearn.model_selection import train_test_split
```

### Split data

```python
# Create a data frame from the csv file
data = pd.read_csv('2.3.1.final_val_ds.csv')

#Remove NULL values
required_cols = ["win%", "performance"]
data = data.dropna(subset=required_cols).copy()

# Split the data into training and testing data
training_data, testing_data = train_test_split(data, test_size=0.25)

# Save the training and testing data to csv files
training_data.to_csv('2.3.1.training_data.csv', index=False)
testing_data.to_csv('../2.4.Model_Testing_and_Evaluation/2.4.1.testing_data.csv', index=False)
```
