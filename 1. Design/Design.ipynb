## Design

#### **Defining Business Problem**

Valorant is a popular game with large-player base that spans globally. With majority of players being stuck in what's defined as 'low-elo' meaning lobbies that mainly have players who are new to the game, excuses and emotions tend to uprise with no proper explanation as to why they are losing. Is it the teammates, the game or you? In Valorant there is a hidden measurement defined as 'MMR' or Match-making Rating this determines the type of teammates and enemies that you are paired against, although there is no proper way to check.

With this opportunity I have identified the potential to create an AI model that is able to assess your MMR based off of your performance. This will allow for a public representation of MMR and for people to be able to identify whether it's a personal issue or their teammates. This model will have to be careful of interpretative bias as players may feel dis-encouraged or misunderstand data, potentially blaming or out-lashing on the community. This should be a small scale model that uses Neural Networks to train patterns based off of user inputs, hopefully allowing for me to remove prior data entries to protect privacy and save computation power.

I've also identified the potential due to the variety of my data to create multiple models comparing my 'Performance' feature against win rate and rank to better understand MMR, thus my final model should be a logistic regression with multiple classes. I also thought about using the model to identify rank as it will be beneficial to users as it can encourage improvement and development. So I'll consider integrating a second model or using rank as a target variable during the process of finding MMR.

By allowing users to predict and interpret their MMR it opens critiques to the game design.

#### **Convert Business Problem into Machine Learning Problem**

In order to convert the business problem, I must first identify what variables I can pull out from the business problem as measurable data. As my goal was to find "MMR" the first goal is to identify a source of data that provides statistics preferably related to "kills", "deaths", "assists", "econ" and "rank", since those were my initial metrics. I want for users to be able to interpret their current ability through concise measurements highlighting areas of improvement. Users should by default be able to input their own data in, not only testing my model but receiving an interpretable and user-friendly response. User-friendly considers interpretative bias and ethical boundaries that I should be wary of before releasing my model for deployment. Ensure that data isn't held within the system to emphasise secure architecture.

#### **Defining Success Metrics**

- In order to most accurately determine MMR due to its secretive nature, I must take into account miniscule factors that can skew data such as statistics about your team and enemy team (unfair matchmaking, hackers) as their capabilities will ultimately affect the outcome of the match. Thus I can presumably best predict MMR based off of "RR" (Ranked Rating) as it scales independent of your team and others, specific to your own performance.

- The hidden mechanic of MMR fluctuates dependent on a various variables and is a universal measurement. An appropriate precaution to take is for data to be from all ranks within the game. As different people receive different teammates, have different habits and capabilities generalising the range of MMR allowing for the model to calculate a better prediction, lowering variance. Another measure to take is to use specifically only “solo queue” data as queueing with friends can affect RR (as if your friend is better you can always queue with them to have an advantage) meaning that I want the fairest data possible with least amount of external interference. This is done to hopefully cut bias of other player's skill, preventing their statistics from interfering with predictions.

- Econ is an optional feature however it takes into account the ability of the user to utilise available weapons and capitalise off of that in game, thus a high econ means a user is able to win with less deaths or worse guns potentially correlating to a player’s skill. This is done to hopefully cut bias off of someone who may have seemingly lower statistics but realistically performed more efficiently.

- I will attempt to feature engineer certain data. Creating my own metrics such as "performance" and "skill" as there’s a clear relationship between certain features and by combining them I can hopefully make more powerful features.

- I can also try to predict MMR through Neural Networks based off of Model of RR, as the relationship may not be as apparent as I make it out to be. This is done to hopefully cut bias from other basic models where Neural Networks can potentially find a more complex and nuanced relationship.

- The best case of action to help classify Ranks and MMR is to convert them into classes through one hot encoding as there more than just one class. As neither targets are numerical, it'll be hard to find high correlation thus KNN, Logistic Regression or Neural networks would be the best approach as data is presented in clusters, thus I can classify my own classes based off of the numerical target provided.
